# Square Demo Notebook

This notebook asks you for a number and then prints the square of that number.

Steps:
1. Run the next cell to confirm the notebook is working.
2. Run the input cell and enter a number when prompted.
3. See the squared result printed below the cell.


```python
print("Notebook is ready")
```

```python
user_input = input("Enter a number: ")


def is_prime(n: int) -> bool:
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    i = 3
    while i * i <= n:
        if n % i == 0:
            return False
        i += 2
    return True


try:
    number = float(user_input)
    squared = number ** 2
    print(f"The square of {number} is {squared}")

    if number.is_integer():
        n_int = int(number)
        if is_prime(n_int):
            print(f"{number} is a prime number.")
        else:
            print(f"{number} is not a prime number.")
    else:
        print(f"{number} is not an integer, so we do not check if it is prime.")
except ValueError:
    print("Please enter a valid numeric value.")
```

```python
import random

n_input = input("How many random numbers should I generate (N)? ")

try:
    n = int(n_input)
    if n <= 0:
        print("Please enter a positive integer for N.")
    else:
        nums = [random.randint(1, 100) for _ in range(n)]
        print(f"Generated {n} random numbers:")
        print(nums)
except ValueError:
    print("Please enter a valid integer for N.")
```
